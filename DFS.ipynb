```python
import queue

def createMaze():
    maze = []
    maze.append(["#","#", "#", "#", "#", "O", "#", "#", "#"])
    maze.append(["#"," ", " ", " ", " ", " ", " ", " ", "#"])
    maze.append(["#"," ", "#", "#", " ", "#", "#", " ", "#"])
    maze.append(["#"," ", "#", " ", " ", " ", "#", " ", "#"])
    maze.append(["#"," ", "#", " ", "#", " ", "#", " ", "#"])
    maze.append(["#"," ", "#", " ", "#", " ", "#", " ", "#"])
    maze.append(["#"," ", "#", " ", "#", " ", "#", "#", "#"])
    maze.append(["#"," ", " ", " ", " ", " ", " ", " ", "#"])
    maze.append(["#","#", "#", "#", "#", "#", "#", "X", "#"])

    return maze


def createMaze3():
    maze =[]
    maze.append([' ', ' ', 'O'])
    maze.append(['#', ' ', ' '])
    maze.append([' ', 'X', ' '])
    return maze
```

```python
def valid(maze,moves):
    for x,y in enumerate(maze[0]):
        if y == 'O':
            start = x
    
    i = start
    j = 0
    for move in moves:
        if move == 'L':
            i-=1
        elif move == 'R':
            i+=1
        elif move == 'U':
            j-=1
        elif move == 'D':
            j+=1
    if i<0 or i > (len(maze[0])-1) or j > (len(maze)-1) or j<0:
        return False
    elif maze[j][i] == '#' or maze[j][i] == '+' or maze[j][i] == 'O':
        return False
    else:
        return True
```

```python
def FindEnd(maze,moves):
    for x,y in enumerate(maze[0]):
        if y == 'O':
            start = x
            break
    
    i = start
    j = 0
    for move in moves:
        if move == 'L':
            i -= 1
        elif move == 'R':
            i += 1
        elif move == 'U':
            j -= 1
        elif move == 'D':
            j += 1
    
    if maze[j][i] == 'X':
        return True
    else:
        return False
```

```python
def fillpath(maze,moves):
    for x,y in enumerate(maze[0]):
        if y == 'O':
            start = x
            break
    
    i = start
    j = 0
    for move in moves:
        if move == 'L':
            i -= 1
            if maze[j][i] == 'X':
                return
            else:
                maze[j][i] = '+'
        elif move == 'R':
            i += 1
            if maze[j][i] == 'X':
                return
            else:
                maze[j][i] = '+'
        elif move == 'U':
            j -= 1
            if maze[j][i] == 'X':
                return
            else:
                maze[j][i] = '+'
        elif move == 'D':
            j += 1
            if maze[j][i] == 'X':
                return
            else:
                maze[j][i] = '+'
```

```python
q = queue.Queue()
q.put('')
maze = createMaze()
move = ''
def run():
    move = q.get()
    if FindEnd(maze,move):
        print("Found path: {}".format(move))
        return
    else:
        for step in ['L','R','U','D']:
            current_step = move + step
            if valid(maze,current_step):
                fillpath(maze,current_step)
                q.put(current_step)
                run()
```
