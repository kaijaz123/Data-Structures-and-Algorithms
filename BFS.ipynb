```python
import queue


def createMaze():
    maze = []
    maze.append(["#","#", "#", "#", "#", "O","#"])
    maze.append(["#"," ", " ", " ", "#", " ","#"])
    maze.append(["#"," ", "#", " ", "#", " ","#"])
    maze.append(["#"," ", "#", " ", " ", " ","#"])
    maze.append(["#"," ", "#", "#", "#", " ","#"])
    maze.append(["#"," ", " ", " ", "#", " ","#"])
    maze.append(["#","#", "#", "#", "#", "X","#"])

    return maze

def createMaze2():
    maze = []
    maze.append(["#","#", "#", "#", "#", "O", "#", "#", "#"])
    maze.append(["#"," ", " ", " ", " ", " ", " ", " ", "#"])
    maze.append(["#"," ", "#", "#", " ", "#", "#", " ", "#"])
    maze.append(["#"," ", "#", " ", " ", " ", "#", " ", "#"])
    maze.append(["#"," ", "#", " ", "#", " ", "#", " ", "#"])
    maze.append(["#"," ", "#", " ", "#", " ", "#", " ", "#"])
    maze.append(["#"," ", "#", " ", "#", " ", "#", "#", "#"])
    maze.append(["#"," ", " ", " ", " ", " ", " ", " ", "#"])
    maze.append(["#","#", "#", "#", "#", "#", "#", "X", "#"])

    return maze

def createMaze3():
    maze =[]
    maze.append([' ', ' ', 'O'])
    maze.append(['#', " ", ' '])
    maze.append([' ', "X", " "])
    return maze
```

```python
def valid(maze,moves):
    for x,y in enumerate(maze[0]):
        if y == 'O':
            start = x
            break
    
    i = start
    j = 0
    for move in moves:
        if move == 'L':
            i -= 1
        elif move == 'R':
            i += 1
        elif move == 'U':
            j -= 1
        elif move == 'D':
            j += 1
    
    if i > (len(maze[0])-1) or i < 0 or j > (len(maze)-1) or j < 0:
        return False
    elif maze[j][i] == '#':
        return False

    return True
        
```

```python
def printMaze(maze,moves):
    for x,y in enumerate(maze[0]):
        if y == 'O':
            start = x
            break
    
    i = start
    j = 0
    for move in moves:
        if move == 'L':
            i -= 1
            if maze[j][i] != 'X':
                maze[j][i] = '+'
        elif move == 'R':
            i += 1
            if maze[j][i] != 'X':
                maze[j][i] = '+'
        elif move == 'U':
            j -= 1
            if maze[j][i] != 'X':
                maze[j][i] = '+'
        elif move == 'D':
            j += 1
            if maze[j][i] != 'X':
                maze[j][i] = '+'
    
    y = 0
    for x in range(len(maze)):
        print(maze[y])
        y += 1
```

```python
def FindEnd(maze,moves):
    for x,y in enumerate(maze[0]):
        if y == 'O':
            start = x
            break
    
    i = start
    j = 0
    for move in moves:
        if move == 'L':
            i -= 1
        elif move == 'R':
            i += 1
        elif move == 'U':
            j -= 1
        elif move == 'D':
            j += 1
            
    if maze[j][i] == 'X':
        print("Found path: {}".format(moves))
        printMaze(maze,moves)
        return True
    else:
        return False
```

```python
maze = createMaze3()
nums = queue.Queue()
nums.put('')
item = ''
while not FindEnd(maze,item):
    item = nums.get()
    print('item: ',item)
    for x in ['L','R','U','D']:
        put = item + x
        if valid(maze,put):
            nums.put(put)
```
