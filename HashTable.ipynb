Linear Probing

```python
class HashTable:
    def __init__(self):
        self.Max = 10
        self.arr = [None for i in range(self.Max)]

    def get_hash(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.Max

    def __setitem__(self, key, val):  
        h = self.get_hash(key)        
        while self.arr[h] is not None:
            if self.arr[h][0] == key:
                self.arr[h] = (key,val)
                break
            else:
                h = (h+1)%self.Max
                
        self.arr[h] = (key,val)
        

    def __getitem__(self, key):       
        h = self.get_hash(key)        
        while self.arr[h][0] != key:
            h = (h+1)%self.Max
        return self.arr[h][1]
            
    def __delitem__(self, key):
        h = self.get_hash(key)
        while self.arr[h][0] != key:
            h = (h+1)%self.Max
        self.arr[h] = None


t = HashTable()
t['march 6'] = 120 
t['march 17'] = 459
t['dec 17'] = 27
t['dec 17'] = 270
t['dec 17']
# del t['march 17']
# t.arr
```

```python
class HashTable:
    def __init__(self):
        self.Max = 10
        self.arr = [None for i in range(self.Max)]

    def get_hash(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.Max
    
    def double_hashing(self,index,key,val):
        found = False
        count = 1
        ori_key = key
        key = self.get_hash(key)
        while self.arr[index] is not None:
            if self.arr[index][0] == ori_key:
                self.arr[index] = (ori_key,val)
                found = True
                break
            else:
                h1 = key%self.Max
                h2 = 7 - (key%7)
                index = (h1 + (count*h2))%self.Max
                count += 1
        
        if found:
            return False
        else:
            return index

    def __setitem__(self, key, val):  
        h = self.get_hash(key)
        if self.arr[h] is None:
            self.arr[h] = (key,val)
            
        elif self.arr[h][0] == key:
            self.arr[h] = (key,val)
                          
        else:
            index = self.double_hashing(h,key,val)
            if index == False:
                return
            else:
                self.arr[index] = (key,val)
        
    def __getitem__(self, key):       
        h = self.get_hash(key)    
        count = 0
        while self.arr[h][0] != key:
            h = ((h+count)^2)%self.Max
            count += 1
        return self.arr[h][1]
            
    def __delitem__(self, key):
        h = self.get_hash(key)
        count = 0
        while self.arr[h][0] != key:
            h = ((h+count)^2)%self.Max
            count += 1
        self.arr[h] = None


t = HashTable()
t['march 6'] = 120 
t['march 17'] = 459
t['dec 17'] = 27
t['march 17'] = 500
# t['march 6'] = 200
# t['march 6']
# t['march 6']
t.arr
```
